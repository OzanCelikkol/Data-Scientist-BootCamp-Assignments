1. Create a blank figure with matplotlib functions. Plot cosine function for the numbers beetwen -3$\pi$ and 3$\pi$ with 0.1 intervals.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot()
plt.show()

cosLabel = np.arange(-3.0,3.0,0.1)
print(cosLabel)
print(np.cos(cosLabel))

plt.plot(cosLabel,np.cos(cosLabel*np.pi))
plt.xlabel("pi")
plt.ylabel("cos")
plt.show()
```

2. This time, create the same plot in red color with object methods. 

```python
import matplotlib.pyplot as plt
import numpy as np

cosLabel = np.arange(-3.0,3.0,0.1)
print(cosLabel)
print(np.cos(cosLabel))

plt.plot(cosLabel,np.cos(cosLabel*np.pi),color="red")
plt.xlabel("pi")
plt.ylabel("cos")
plt.show()
```
