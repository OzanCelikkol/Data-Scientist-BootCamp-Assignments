1. Plot a bar chart for total cases of the 20 countries that have biggest numbers.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)

sumDF = pd.DataFrame(df.groupby("location")["new_cases"].sum())
sumDF = (sumDF["new_cases"].sort_values()).tail(21)
sumDF = sumDF[sumDF.index != "World"]
display(sumDF)
plt.figure(figsize =(30,10))
plt.bar(sumDF.index,sumDF)
plt.show()
```

2. Plot a histogram for daily deaths for any country you choose. Make three subplots for different bins.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
estoniaDF = df[df.location == "United States"]
estoniaDailyDeaths = estoniaDF.loc[:,"new_deaths"]
display(estoniaDF)

plt.subplot(3,1,1)
plt.title("Daily Deaths in Estonia", fontsize = 20)
plt.xlabel("Days", fontsize = 10)
plt.ylabel("Daily Deaths", fontsize = 10)
plt.hist(estoniaDailyDeaths, bins = 50, color = "yellow")
plt.show()

plt.subplot(3,1,2)
plt.title("Daily Deaths in Estonia", fontsize = 20)
plt.xlabel("Days", fontsize = 10)
plt.ylabel("Daily Deaths", fontsize = 10)
plt.hist(estoniaDailyDeaths, bins = 30, color = "yellow")
plt.show()

plt.subplot(3,1,2)
plt.title("Daily Deaths in Estonia", fontsize = 20)
plt.xlabel("Days", fontsize = 10)
plt.ylabel("Daily Deaths", fontsize = 10)
plt.hist(estoniaDailyDeaths, bins = 15, color = "yellow")
plt.show()
```

3. Plot a scatter plot of new cases and new death for Germany and France.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
germanyFranceDF = pd.DataFrame(df[df["location"].isin(["Germany","France"])])
display(germanyFranceDF)

plt.figure(figsize = (15,5))
plt.xlabel("New Cases", fontsize = 30)
plt.ylabel("New Deaths", fontsize = 30)
display(germanyFranceDF.groupby(by = "location"))
for location in ["Germany","France"]:
    plt.scatter(germanyFranceDF[germanyFranceDF.location == location].new_cases ,
            germanyFranceDF[germanyFranceDF.location == location].new_deaths,
            label = location)
plt.legend()
plt.show()
```

4. Plot a boxplot for daily deaths for any country you choose.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
swedenDF = df[df["location"] == "Sweden"]
swedenDailyDeaths = swedenDF.loc[:,"new_deaths"]

plt.figure(figsize = (15,5))
plt.title("Daily Deaths in Sweden", fontsize = 30)
plt.xlabel("Estonia", fontsize = 10)
plt.ylabel("New Deaths", fontsize = 10)
plt.boxplot(swedenDailyDeaths.dropna())
plt.show()
```

5. Calculate the total case for each continent and plot a pie chart 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
totalCaseContinent = pd.DataFrame([df.groupby(by = "continent")["new_cases"].sum()])
display(totalCaseContinent)
print(totalCaseContinent.shape)
plt.figure(figsize = (15,15))
plt.title("Total Cases by Continents")
plt.pie(totalCaseContinent.loc["new_cases",:], labels = totalCaseContinent.columns.values, autopct='%1.1f%%',
        shadow=False, startangle=90)
plt.show()
```
