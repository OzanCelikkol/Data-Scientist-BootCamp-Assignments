1. Plot a line plot with seaborn for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns

sns.set()
df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
display(df)
totalDeaths = df[df["location"].isin(["Spain","France","Germany","Italy"])]
totalDeaths = totalDeaths[totalDeaths["date"] > dt.datetime(2020,4,1)]
display(totalDeaths)

plt.figure(figsize = (15,15))
plt.title("Total Deaths By Countries", fontsize = 50)
plt.xlabel("Date", fontsize = 30)
plt.ylabel("Total Deaths", fontsize = 30)
plt.plot(totalDeaths[totalDeaths["location"].isin(["Spain"])]["date"],
         totalDeaths[totalDeaths["location"].isin(["Spain"])]["total_deaths"], color = "yellow", lw = 5)
plt.plot(totalDeaths[totalDeaths["location"].isin(["France"])]["date"],
         totalDeaths[totalDeaths["location"].isin(["France"])]["total_deaths"], color = "blue", lw = 5)
plt.plot(totalDeaths[totalDeaths["location"].isin(["Germany"])]["date"],
         totalDeaths[totalDeaths["location"].isin(["Germany"])]["total_deaths"], color = "red", lw = 5)
plt.plot(totalDeaths[totalDeaths["location"].isin(["Italy"])]["date"],
         totalDeaths[totalDeaths["location"].isin(["Italy"])]["total_deaths"], color = "green", lw = 5)
plt.legend(["Spain","France","Germany","Italy"])
plt.show()
```

2. Plot a bar plot with seaborn for average death number that compares continents.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns

sns.set()
df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
display(df)
continent = df.groupby(by = "continent").agg({"new_deaths":["mean"]})
display(continent)

plt.figure(figsize = (20,5))
plt.title("Average Total Deaths By Continent", fontsize = 50)
plt.ylabel("Average Total Deaths", fontsize = 20)
plt.xlabel("Continents", fontsize = 20)
sns.barplot(x = continent.index, y = continent.iloc[:,0][:], data = continent)
plt.show()
```

3. Plot a histogram for daily deaths for any country you choose. Make four subplots for different bin counts and `kde` arguments.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns

sns.set()
df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
dailyDeaths = df[df["location"] == "United States"]
display(dailyDeaths)

plt.subplot(4,1,1)
plt.figure(figsize = (15,5))
plt.title("Daily Deaths In USA", fontsize = 50)
plt.xlabel("Date", fontsize = 30)
plt.ylabel("Total Deaths in USA", fontsize = 30)
sns.distplot(dailyDeaths.new_deaths.dropna(), kde = False, bins = 50)
plt.show()

plt.subplot(4,1,2)
plt.figure(figsize = (15,5))
plt.ylabel("Total Deaths in USA", fontsize = 30)
sns.distplot(dailyDeaths.new_deaths.dropna(), kde = True, bins = 50)
plt.show()

plt.subplot(4,1,3)
plt.figure(figsize = (15,5))
plt.ylabel("Total Deaths in USA", fontsize = 30)
sns.distplot(dailyDeaths.new_deaths.dropna(), kde = False, bins = 25)
plt.show()

plt.subplot(4,1,4)
plt.figure(figsize = (15,5))
plt.ylabel("Total Deaths in USA", fontsize = 30)
sns.distplot(dailyDeaths.new_deaths.dropna(), kde = True, bins = 25)
plt.show()
```

4. Create a figure and three subplots containing boxplot, violin plot and swarm plot for daily deaths of two countries you choose.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
import seaborn as sns

sns.set()
df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
chnusaDF = df[df["location"].isin(["China","United States"])]

plt.subplot(1,3,1)
plt.figure(figsize = (15,5))
plt.title("Daily Detahs", fontsize = 50)
sns.barplot(y = "new_deaths",x = "location", data = chnusaDF, palette = "ocean")
plt.show()

plt.subplot(1,3,2)
plt.figure(figsize = (15,5))
sns.violinplot(x = "location", y = "new_deaths", data = chnusaDF)
plt.show()

plt.subplot(1,3,3)
plt.figure(figsize = (15,5))
sns.swarmplot(x = "location", y = "new_deaths", data = chnusaDF)
plt.show()
```
