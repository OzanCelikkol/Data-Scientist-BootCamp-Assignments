```python
import pandas as pd
df = pd.read_csv("train.csv")
display(df)
def1 = df[(df['Survived']==1) & (df['Age']<30)]
display(def1)
survRate = def1.shape[0] / df.shape[0]
print("survival rate for passengers whose ages are under 30 = % {:.5f}".format(survRate*100))
dfFemale = df[(df['Survived']==1) & (df['Sex'] == 'female')]
display(dfFemale)
survRateFemale = dfFemale.shape[0] / df.shape[0]
print("survival rate for female passengers = % {:.5f}".format(survRateFemale*100))
dfMale = df[(df['Survived']==1) & (df['Sex'] == 'male')]
display(dfMale)
survRateMale = dfMale.shape[0] / df.shape[0]
print("survival rate for male passengers = % {:.5f}".format(survRateMale*100))
```
