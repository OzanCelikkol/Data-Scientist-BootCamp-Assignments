In this assignment, you will make some plots about Corona. First, download the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data).

- First, filter the data for four countries (Spain, France, Germany, Italy) and the last 100 days. 

- Create a figure with two subplots. The left subplot will show total cases, and the right one will show total deaths for these countries

- Set titles, X and Y labels, change the font, font size, and font color for each subplot.

- Change color, width, and type of lines.

- Put a legend to bottom right.

- Change the color and rotation of the X-axis ticks.

- And set a title for the figure (not for the plots).

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta


df = pd.read_csv('owid-covid-data.csv', parse_dates=["date"], low_memory=False)
newdf = df["date"].sort_values()
lastDay = newdf.iloc[-1]
theDay = lastDay - timedelta(days = 100)
df1 = pd.DataFrame(df[df["location"].isin(["Spain","France","Germany","Italy"]) & (df["date"] >= theDay)])
display(df1)
plt.figure(figsize = (25,10))
plt.subplot(2,2,1)
plt.xlabel("Date", fontsize = 8, fontstyle = "oblique", color = "black")
plt.ylabel("Total Cases", fontsize = 8, fontstyle = "oblique", color = "black")
plt.title("TOTAL CASES")
plt.xticks(rotation = 45, color = "purple")
plt.plot(df1[df1["location"].isin(["Germany"])]["date"],df1[df1["location"].isin(["Germany"])]["total_cases"], c ="red", lw = 5,
         marker = "+")
plt.plot(df1[df1["location"].isin(["France"])]["date"],df1[df1["location"].isin(["France"])]["total_cases"], c = "blue", lw = 4,
         marker = "o")
plt.plot(df1[df1["location"].isin(["Spain"])]["date"],df1[df1["location"].isin(["Spain"])]["total_cases"], c = "yellow", lw = 3,
         marker = "x")
plt.plot(df1[df1["location"].isin(["Italy"])]["date"],df1[df1["location"].isin(["Italy"])]["total_cases"], c = "green", lw = 2,
         marker= "*")
plt.legend(["GERMANY","FRANCE","SPAIN","ITALY"], loc = 4)
plt.show()
plt.figure(figsize = (25,10))
plt.subplot(2,2,2)
plt.xlabel("Date", fontsize = 8, fontstyle = "italic", color = "black")
plt.ylabel("Total Deaths", fontsize = 8, fontstyle = "italic", color = "black")
plt.title("TOTAL DEATHS")
plt.xticks(rotation = 45, color = "purple")
plt.plot(df1[df1["location"].isin(["Germany"])]["date"],df1[df1["location"].isin(["Germany"])]["total_deaths"], c = "red", lw = 5,
        marker = "+")
plt.plot(df1[df1["location"].isin(["France"])]["date"],df1[df1["location"].isin(["France"])]["total_deaths"], c = "blue", lw = 4,
        marker = "o")
plt.plot(df1[df1["location"].isin(["Spain"])]["date"],df1[df1["location"].isin(["Spain"])]["total_deaths"], c = "yellow", lw = 3,
        marker = "x")
plt.plot(df1[df1["location"].isin(["Italy"])]["date"],df1[df1["location"].isin(["Italy"])]["total_deaths"], c = "green", lw = 2,
        marker = "*")
plt.legend(["GERMANY","FRANCE","SPAIN","ITALY"], loc = 4)
plt.show()
```
