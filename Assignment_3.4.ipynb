In this assignment you will continue to make some plots on the [Coronavirus Source Data](https://ourworldindata.org/coronavirus-source-data). For plotting you will use Plotly Express. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

df = pd.read_csv('../1.1-DataSets/owid-covid-data.csv', parse_dates=["date"], low_memory=False)
```

1. Plot a line plot with plotly express for total deaths four the four countries (Spain, France, Germany, Italy) after April 1, 2020.

```python
df1 = df[df["location"].isin(["Spain","France","Germany","Italy"])]
display(df1)
fig = px.line(df1, x="date", y = "total_deaths", title = "Total Deaths By Countries",color = "location")
fig.show()
```

2. Plot a bar plot with plotly express for average death number that compares continents.

```python
df2 = pd.DataFrame([df.groupby("continent")["new_deaths"].mean()])
df2 = df2.T
display(df2)
fig = px.bar(data_frame = df2, x = df2.index, y = "new_deaths", color = df2.index)
fig.show()
```

3. Plot a histogram for daily deaths for any country you choose. 

```python
df3 = df[df["location"] == "United States"]
display(df3)
fig = px.histogram(df3, x = "new_deaths", title = "Daily Deaths in USA")
fig.show()
```
